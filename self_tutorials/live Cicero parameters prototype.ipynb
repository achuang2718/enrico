```python
def update_run(bc, run_id, params):
    if isinstance(run_id, float):
        run_id = int(run_id)
    response = bc._send_message('get','/runs/' + str(run_id) + '/').json()
    response.update(params)
    new_response = bc._send_message('put','/runs/' + str(run_id) + '/', data = json.dumps(response))
    if new_response.status_code == 200:
        print('updated run ' + str(run_id) + ' successfully')
    else:
        print('error: updating run' + str(run_id) + ' failed')
        
params  = {'notes': {'labbook':'alex was here','bad shot': False}}
update_run(bc, 191372, params)
response = bc._send_message('get','/runs/191372/').json()
response
```
